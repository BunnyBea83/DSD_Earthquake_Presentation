```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# read in both CSV files
eq_95_23 = pd.read_csv("data/earthquake_1995-2023.csv")
eq_data = pd.read_csv("data/earthquake_data.csv")
```

```python
# EDA

eq_95_23.info()
```

```python
#EDA

eq_data.info()
```

```python
# Data Cleaning and Transformation 

eq_95_23 = eq_95_23.drop(['alert','continent', 'country'], axis=1)
eq_95_23
```

```python
# Data Cleaning and Transformation 

eq_data = eq_data.drop(['alert','continent', 'country'], axis=1)
eq_data
```

```python
# Data joining

""" 
You can see there are many duplicates between the two CSV files. Before dropping duplicates 
there were 1782 rows and 998 after. 
"""

combined = pd.concat([eq_95_23,eq_data])
combined = combined[['date_time','depth', 'magnitude']]
display("Before dropping duplicates")
display(combined)
display("-------------------------------------------------------")
display("After dropping duplicates")
combined = combined.drop_duplicates()
display(combined)
```

```python
# Data Visualization

eq_95_23_depth = eq_95_23['depth']
eq_95_23_magnitude = eq_95_23['magnitude']
plt.figure(figsize=(11, 6))
plt.scatter(eq_95_23_depth, eq_95_23_magnitude, color='red')
plt.grid()
plt.suptitle('Earthquake Depth vs. Magnitude (1995–2023)', fontsize=20)
plt.title('eq_95_23.csv')
plt.xlabel('Depth (km)', fontsize=15)
plt.ylabel('Magnitude', fontsize=15)
plt.show()
```

```python
# Data Visualization

eq_data_depth = eq_data['depth']
eq_data_magnitude = eq_data['magnitude']
plt.figure(figsize=(11, 6))
plt.scatter(eq_data_depth, eq_data_magnitude, color='red')
plt.grid()
plt.suptitle('Earthquake Depth vs. Magnitude (2001–2022)', fontsize=20)
plt.title('eq_data.csv')
plt.xlabel('Depth (km)', fontsize=15)
plt.ylabel('Magnitude', fontsize=15)
plt.show()
```

```python
# Data Visualization

combined_depth = combined['depth']
combined_magnitude = combined['magnitude']
plt.figure(figsize=(11, 6))
plt.scatter(combined_depth, combined_magnitude, color='red')
plt.grid()
plt.suptitle('Earthquake Depth vs. Magnitude (Combined Datasets)', fontsize=20)
plt.title('eq_95_23.csv and eq_data.csv')
plt.xlabel('Depth (km)', fontsize=15)
plt.ylabel('Magnitude', fontsize=15)
plt.show()
```
